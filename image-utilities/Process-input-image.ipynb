```python
def contour_draw(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    gray_plot = plt.figure(1)
    plt.imshow(img)

    edge_plot = plt.figure(2)
    edges = cv2.Canny(gray_img, threshold1=30, threshold2=100)
    plt.imshow(edges, cmap="gray")

    print(variance_of_laplacian(gray_img))
    
    thres_plot = plt.figure(3)
    ret, thresh = cv2.threshold(gray_img, 64, 255, 0)
#     thresh = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 7, 2)
    plt.imshow(thresh, cmap="gray")
    # contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Finding Contours
    # Use a copy of the image e.g. edged.copy()
    # since findContours alters the image
    _, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    print("Number of Contours found = " + str(len(contours)))

    max_contour = max(contours, key = cv2.contourArea)
    x,y,w,h = cv2.boundingRect(max_contour)

    contour_plot = plt.figure(4)
    cv2.drawContours(gray_img, contours, -1, (255, 0, 0), 3)
    plt.imshow(gray_img, cmap="gray")

    max_contour_plot = plt.figure(5)
    cv2.rectangle(gray_img, (x,y), (x+w, y+h), (0, 255, 0), 5)
    
    mval = min(w, h)
    cv2.rectangle(gray_img, (math.floor(x + w/2 - mval/2), math.floor(y + h/2 - mval/2)), (math.floor(x + w/2 + mval/2), math.floor(y + h/2 + mval/2)), (0, 0, 255), 2)
    
    plt.imshow(gray_img, cmap="gray")

    # print(contours)

    plt.show()
```

```python
def crop_img(file_name, img):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray_img, 64, 255, 0)
    _, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) == 0:
        print("find no contours for file: " + file_name)
        return None
    max_contour = max(contours, key = cv2.contourArea)
    x,y,w,h = cv2.boundingRect(max_contour)
    return img[y:y+h, x:x+w]
```

```python
def crop_img_round(file_name, img):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray_img, 64, 255, 0)
    _, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) == 0:
        print("find no contours for file: " + file_name)
        return None
    max_contour = max(contours, key = cv2.contourArea)
    x,y,w,h = cv2.boundingRect(max_contour)
    mval = min(w, h)
    newy = math.floor(y + h/2 - mval/2)
    newx = math.floor(x + w/2 - mval/2)
    return img[newy:newy+mval, newx:newx+mval]
```

```python
from_path = "/Users/phunp/Documents/Flinders/Thesis/data/real_data/extract_data_ori/"
to_path = "/Users/phunp/Documents/Flinders/Thesis/data/real_data/extract_data_crop_round/"

count = 0
no_contours = 0
errors = 0
for filename in os.listdir(from_path):
    img = cv2.imread(os.path.join(from_path,filename))
    try :
        crop = crop_img_round(filename, img)
        if crop is not None:
            cv2.imwrite(to_path + filename, crop)
            count += 1
        else:
            no_contours += 1
    except:
        print("Something wrong with file: " + filename)
        errors += 1
    if count % 500 == 0:
        print("process %d file" % count)
    
print("total process success %d, no contours found %d, errors %d" % (count, no_contours, errors))
```
